## Задача: 

Тело массой $m = 1 кг$ брошено c поверхности земли вертикально вверх с начальной скоростью $U_{0} = 19,6 м/с$. Определить изменение потенциальной энергии тела за промежуток времени $t = 2 с$ после броска.
Построить графики зависимости $U_{y}(t) и Е_{p}(t)$. Сопротивление воздуха не учитывать.

Скорость $U$ тела изменяется так:

  $U=U_{0}-g*t$

Нетрудно заметить, что по истечении двух секунд тело будет находиться в точке наивысшего подъема, поэтому

  $h=U_{0}^2/2*g$

А потенциальная энергия изменится на

  $E_{p}=m*g*h=m*U_{0}^2/2=19,6^2/2=192$

Ответ: энергия изменится на $192 Дж$.
